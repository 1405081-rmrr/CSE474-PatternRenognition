```python
import numpy as np
from scipy.stats import multivariate_normal
file=open('train.txt','r')
data=file.read()

a = 1.0
b = 0.5

def noise():
    mu, sigma = 0.0,20   #mean and standard deviation
    s = np.random.normal(mu, sigma, 1)
    return s
    
def funcN(cur, prev):
    return a*cur + b*prev +noise()  #The equation is xk-1=0.5Ik-1+Ik-2 a=1 ; b=0.5
def func(cur, prev):
    return a*cur + b* prev
mu = []
cnt = [] 
sigma = []
for i in range(8):
    mu.append([0, 0])
    #cnt.append([0, 0])
    sigma.append([[0, 0], [0, 0]])


mu =np.array(mu)
#cnt =np.array(cnt)
sigma = np.array(sigma)
#print(sigma.shape)

def train():
    global mu, cnt
    for i in range(len(data)-2):
        a1 = int(data[i+2])
        b1 = int(data[i+1])
        c1 = int(data[i])
        xk = funcN(a1, b1)
        xk_prev = funcN(b1, c1)
        index = 4*a1+2*b1+c1 #binary to decimal convertion
        mu[index][0] += xk
        mu[index][1] += xk_prev
        #cnt[index][0] += 1
        #cnt[index][1] += 1

train()

def test(): #Euclidean distance
    tot = 0
    file=open('test.txt','r')
    data=file.read()
    global mu, cnt
    for i in range(len(data)-2):
        a1 = int(data[i+2])
        b1 = int(data[i+1])
        c1 = int(data[i])
        xk = func(a1, b1)
        xk_prev = func(b1, c1)
        #index = 4*a1+2*b1+c1
        dis = 1e5
        idx = -1
        pred = -1
        for j in range(8):
            if((mu[j][0]-xk)**2 + (mu[j][1]-xk_prev)**2 < dis):
                dis = (mu[j][0]-xk)**2 + (mu[j][1]-xk_prev)**2
                idx = j
        if(idx<=3):
            pred = 0
        else:
            pred = 1
        if(pred != a1):
            print("mistake at ", i)
            tot+=1
    print("total mistake = ", tot)
test()
```
