```python
import numpy as np
import pandas as pd
import sklearn
from collections import Counter
import collections
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
class Perceptron:
    def __init__(self, learning_rate=0.1, iteration=100):
        self.lr = learning_rate
        self.iteration = iteration
        self.activation_func = self._unit_step_func
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        number_of_samples, number_of_features = X.shape

        # init parameters
        self.weights = np.zeros(number_of_features)
        self.bias = 0
        y_ = np.array(y)
        #print("Actual y ",y_)

        for _ in range(self.iteration):

            for idx, x_i in enumerate(X):
                #print(x_i)
                #print(self.weights)
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_func(linear_output)
                #print("In Loop ",y_predicted)
                # Perceptron update rule
                update = self.lr * (y_[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update
                #print(linear_output)
                
            

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.activation_func(linear_output)
        #print("Predicted Value Count of Y ",collections.Counter(y_predicted))
        #print(y_predicted)
        return y_predicted

    def _unit_step_func(self, x):
        x=abs(x)
        return np.where(x <2,1,2)
    def Pocket(self,X_train,y_train,X_test,y_test,learningRate):
        number_of_samples, number_of_features = X_train.shape
        self.weights=np.zeros(number_of_features)
        self.weights = np.reshape(self.weights, (-1, 1))
        misClassifications = 1
        classification=1
        minMisclassifications = 1000
        iteration = 0
        while (misClassifications != 0 and classification!=0 and (iteration<1000)):
            iteration += 1
            misClassifications = 0
            classification=0
            for i in range(0, len(X_train)):
                currentX_train = X_train[i].reshape(-1, X_train.shape[1])
                currentX_test = X_test[i].reshape(-1, X_test.shape[1])
                currentY_train = y_train[i]
                currentY_test = y_test[i]
                wTx = np.dot(currentX_train, self.weights)[0][0]
                
                if currentY_train == 2 and wTx <= 0:
                    misClassifications += 1
                    #self.weights = self.weights + learningRate * np.transpose(currentX_train)
                    self.weights = self.weights + learningRate * np.transpose(currentX_train)
                elif currentY_train == 1 and wTx > 0:
                    misClassifications += 1
                    #self.weights = self.weights - learningRate * np.transpose(currentX_train)
                    self.weights = self.weights - learningRate * np.transpose(currentX_train)

    
            if misClassifications<minMisclassifications:
                minMisclassifications = misClassifications
                self.updated_weight=self.weights

        output=np.dot(X_test,self.updated_weight)
        y_predicted = self.activation_func(output)
        y_predicted = y_predicted.reshape([1, X_train.shape[0]])
        return y_predicted
    def RP(self,X_train,y_train,X_test,y_test,learningRate):
        number_of_samples, number_of_features = X_train.shape
        self.weights=np.zeros(number_of_features)
        self.weights = np.reshape(self.weights, (-1, 1))
        misClassifications = 1
        classification=1
        minMisclassifications = 1000
        iteration = 0
        while (misClassifications != 0 and classification!=0 and (iteration<1000)):
            iteration += 1
            misClassifications = 0
            classification=0
            for i in range(0, len(X_train)):
                currentX_train = X_train[i].reshape(-1, X_train.shape[1])
                currentX_test = X_test[i].reshape(-1, X_test.shape[1])
                currentY_train = y_train[i]
                currentY_test = y_test[i]
                wTx = np.dot(currentX_train, self.weights)[0][0]
                
                if currentY_train == 2 and wTx <= 0:
                    misClassifications += 1
                    #self.weights = self.weights + learningRate * np.transpose(currentX_train)
                    self.weights = self.weights + learningRate * np.transpose(currentX_train)
                elif currentY_train == 1 and wTx > 0:
                    misClassifications += 1
                    #self.weights = self.weights - learningRate * np.transpose(currentX_train)
                    self.weights = self.weights - learningRate * np.transpose(currentX_train)

        output=np.dot(X_test,self.weights)
        y_predicted = self.activation_func(output)
        y_predicted = y_predicted.reshape([1, X_train.shape[0]])
        return y_predicted
    
if __name__ == "__main__":
    # Imports
    import matplotlib.pyplot as plt
    from sklearn.model_selection import train_test_split
    from collections import Counter
    import collections
    
    def fileReadSeparable():
        feature=list()
        label=list()
        flag=0
        f=open("trainLinearlySeparable.txt")
        for line in f:
            info=list()
            x=line.split()
            if(flag==0):
                for item in x:
                    item=float(item)
                    info.append(item)
                flag=1
            elif(flag==1):
                for i in range(0,len(x)-1):
                    item=float(x[i])
                    info.append(item)
                feature.append(info)
                label.append(float(x[len(x)-1]))
        X=np.array(feature)
        y=np.array(label)
        f.close()
        feature=list()
        label=list()
        flag=0
        f=open("testLinearlySeparable.txt")
        for line in f:
            info=list()
            x=line.split()
            for i in range(0,len(x)-1):
                item=float(x[i])
                info.append(item)
            feature.append(info)
            label.append(float(x[len(x)-1]))
        X_test=np.array(feature)
        y_test=np.array(label)
        f.close()
        X_train, X_test, y_train, y_test =(X,X_test,y,y_test)
        #print(X_train)
        return X_train,y_train,X_test,y_test
    def fileReadNonSeparable():
        feature=list()
        label=list()
        flag=0
        f=open("trainLinearlyNonSeparable.txt")
        for line in f:
            info=list()
            x=line.split()
            if(flag==0):
                for item in x:
                    item=float(item)
                    info.append(item)
                flag=1
            elif(flag==1):
                for i in range(0,len(x)-1):
                    item=float(x[i])
                    info.append(item)
                feature.append(info)
                label.append(float(x[len(x)-1]))
        X=np.array(feature)
        y=np.array(label)
        f.close()
        feature=list()
        label=list()
        flag=0
        f=open("testLinearlyNonSeparable.txt")
        for line in f:
            info=list()
            x=line.split()
            for i in range(0,len(x)-1):
                item=float(x[i])
                info.append(item)
            feature.append(info)
            label.append(float(x[len(x)-1]))
        X_test=np.array(feature)
        y_test=np.array(label)
        f.close()
        X_train, X_test, y_train, y_test =(X,X_test,y,y_test)
        #print(X_train)
        return X_train,y_train,X_test,y_test
    
    X_train,y_train,X_test,y_test=fileReadSeparable()
    X_train_non,y_train_non,X_test_non,y_test_non=fileReadNonSeparable()
    p = Perceptron(learning_rate=0.1, iteration=1000)
    p.fit(X_train, y_train)
    predictions_basic = p.predict(X_test)
    print('Basic Perceptron Accuracy: %.3f' % accuracy_score(y_test,predictions_basic))
    p=Perceptron()
    predictions_pocket=p.Pocket(X_train_non,y_train_non,X_test_non,y_test_non,.01)
    predictions_pocket=predictions_pocket.ravel()
    print("Pocket Perceptron classification accuracy %.3f" % accuracy_score(y_test_non, predictions_pocket))
    #print(predictions_pocket)
    predictions_rp=p.RP(X_train_non,y_train_non,X_test_non,y_test_non,.01)
    predictions_rp=predictions_rp.ravel()
    #print("Pre RP ",predictions_rp)
    print("Reward Punishment Perceptron classification accuracy %.3f" % accuracy_score(y_test_non, predictions_rp))
    
    #File Update
    header=["F1","F2","F3","F4","Actual Class","Predicted Class"]
    read_file = pd.read_csv (r'testLinearlyNonSeparable.txt')
    x=read_file.to_csv (r'testLinearlyNonSeparable_pocket.csv', index=None)
    df=pd.read_csv("testLinearlyNonSeparable_pocket.csv",sep='\s+',header=None)
    value=0
    df["5"] = ""
    df["5"]=predictions_pocket
    print(df)
    df=df.to_csv("testLinearlyNonSeparable_pocket.csv",index=False,header=["F1","F2","F3","F4","F5","Actual Class","Predicted Class"])
    
    read_file = pd.read_csv (r'testLinearlyNonSeparable.txt')
    x=read_file.to_csv (r'testLinearlyNonSeparable_rp.csv', index=None)
    df=pd.read_csv("testLinearlyNonSeparable_rp.csv",sep='\s+',header=None)
    value=0
    df["5"] = ""
    df["5"]=predictions_rp
    print(df)
    df=df.to_csv("testLinearlyNonSeparable_rp.csv",index=False,header=["F1","F2","F3","F4","F5","Actual Class","Predicted Class"])
    
    read_file = pd.read_csv (r'testLinearlySeparable.txt')
    x=read_file.to_csv (r'testLinearlySeparable_basic.csv', index=None)
    df=pd.read_csv("testLinearlySeparable_basic.csv",sep='\s+',header=None)
    value=0
    df["5"] = ""
    df["5"]=predictions_basic
    print(df)
    df=df.to_csv("testLinearlySeparable_basic.csv",index=False,header=["F1","F2","F3","F4","F5","Actual Class","Predicted Class"])
    
```
